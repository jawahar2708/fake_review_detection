```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
import joblib
from tqdm import tqdm
tqdm.pandas()

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
```

```python
# Load your Excel files
main_path = "/content/Subscriptrion box.xlsx"
meta_path = "/content/subscription box-meta.xlsx"

excel_main = pd.ExcelFile(main_path)
excel_meta = pd.ExcelFile(meta_path)

print("Main file sheets:", excel_main.sheet_names)
print("Meta file sheets:", excel_meta.sheet_names)

# Preview the first few rows of each sheet
for s in excel_main.sheet_names:
    print("\n---", s, "---")
    display(pd.read_excel(main_path, sheet_name=s).head())

for s in excel_meta.sheet_names:
    print("\n---", s, "---")
    display(pd.read_excel(meta_path, sheet_name=s).head())
```

```python
reviews_sheet = "Sheet1"     # change if your sheet name is different
meta_sheet = "Sheet1"        # change if your sheet name is different

df_reviews = pd.read_excel(main_path, sheet_name=reviews_sheet)
df_meta = pd.read_excel(meta_path, sheet_name=meta_sheet)

print("Reviews shape:", df_reviews.shape)
print("Meta shape:", df_meta.shape)
df_reviews.head()
```

```python
# Inspect columns
print("Review columns:", df_reviews.columns.tolist())
print("Meta columns:", df_meta.columns.tolist())

# Detect likely key and text columns
possible_text = ['review_text', 'text', 'review', 'comment', 'body']
possible_user = ['user_id', 'reviewer_id', 'author', 'userid']
possible_product = ['product_id', 'asin', 'item_id']
possible_rating = ['rating', 'stars', 'score']

def find_first(df, options):
    for c in df.columns:
        if c.lower() in [x.lower() for x in options]:
            return c
    return None

text_col = find_first(df_reviews, possible_text)
user_col = find_first(df_reviews, possible_user) or find_first(df_meta, possible_user)
product_col = find_first(df_reviews, possible_product)
rating_col = find_first(df_reviews, possible_rating)

print("Detected text_col:", text_col)
print("Detected user_col:", user_col)
print("Detected product_col:", product_col)
print("Detected rating_col:", rating_col)
```

```python
# Merge reviews and meta datasets if a common key is found
if user_col and user_col in df_reviews.columns and user_col in df_meta.columns:
    df = df_reviews.merge(df_meta, on=user_col, how='left', suffixes=('', '_meta'))
elif product_col and product_col in df_reviews.columns and product_col in df_meta.columns:
    df = df_reviews.merge(df_meta, on=product_col, how='left', suffixes=('', '_meta'))
else:
    print("⚠️ No common key found — using reviews dataset only")
    df = df_reviews.copy()

print("Merged shape:", df.shape)
df.head()
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))

def clean_text(s):
    if pd.isna(s):
        return ""
    s = s.lower()
    s = re.sub(r"http\S+", "", s)
    s = re.sub(r"[^a-z0-9\s]", " ", s)
    s = re.sub(r"\s+", " ", s).strip()
    return s

df['clean_text'] = df[text_col].astype(str).progress_apply(clean_text)
df.head()
```

```python
df['review_length'] = df['clean_text'].apply(lambda x: len(x.split()))
df['char_length'] = df['clean_text'].apply(len)
df['exclamation_count'] = df[text_col].astype(str).apply(lambda x: x.count('!'))
df['uppercase_ratio'] = df[text_col].astype(str).apply(lambda x: sum(1 for ch in x if ch.isupper()) / (len(x)+1))
df.head()
```

```python
label_col = None
for c in df.columns:
    if c.lower() in ['label', 'fake', 'is_fake', 'target']:
        label_col = c
        break

if label_col:
    print("✅ Found label column:", label_col)
    df['label'] = df[label_col].astype(int)
else:
    print("⚠️ No label found — creating heuristic labels (temporary)")
    df['label'] = np.where((df['review_length'] < 5) & (df.get(rating_col, 5) >= 5), 1, 0)

df['label'].value_counts()
```

```python
train_df, test_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=RANDOM_SEED)
train_df, val_df = train_test_split(train_df, test_size=0.15, stratify=train_df['label'], random_state=RANDOM_SEED)

print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))
```

```python
tfidf = TfidfVectorizer(max_features=20000, ngram_range=(1,2))
X_train = tfidf.fit_transform(train_df['clean_text'])
X_val = tfidf.transform(val_df['clean_text'])
X_test = tfidf.transform(test_df['clean_text'])

y_train, y_val, y_test = train_df['label'], val_df['label'], test_df['label']

clf = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=RANDOM_SEED)
clf.fit(X_train, y_train)

val_pred = clf.predict(X_val)
test_pred = clf.predict(X_test)

print("Validation Report:\n", classification_report(y_val, val_pred))
print("Test Report:\n", classification_report(y_test, test_pred))
print("ROC-AUC:", roc_auc_score(y_test, clf.predict_proba(X_test)[:,1]))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import pandas as pd

results = []

def add_result(name, y_true, y_pred, y_prob=None):
    acc = accuracy_score(y_true, y_pred)
    pre = precision_score(y_true, y_pred)
    rec = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_prob) if y_prob is not None else None
    results.append({'Model': name, 'Accuracy': acc, 'Precision': pre, 'Recall': rec, 'F1-Score': f1, 'ROC-AUC': auc})

add_result('Logistic Regression', y_test, test_pred, clf.predict_proba(X_test)[:,1])
add_result('XGBoost', test_df['label'], pred_xgb, proba_xgb)
add_result('Naive Bayes', test_df['label'], pred_nb, None)
add_result('SVM', test_df['label'], pred_svm, None)
add_result('Random Forest', test_df['label'], pred_rf, proba_rf)


results_df = pd.DataFrame(results)
results_df
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from xgboost import XGBClassifier
from scipy.sparse import hstack
import joblib, json

# ---------------------
# Load & prepare data
# ---------------------

df['clean_text'] = df['clean_text'].astype(str)
df['review_length'] = df['clean_text'].apply(lambda x: len(x.split()))
df['char_length'] = df['clean_text'].apply(lambda x: len(x))
df['exclamation_count'] = df['clean_text'].apply(lambda x: x.count('!'))
df['uppercase_ratio'] = df['clean_text'].apply(lambda x: sum(1 for c in x if c.isupper()) / max(len(x), 1))

numeric_cols = ['review_length', 'char_length', 'exclamation_count', 'uppercase_ratio']

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])

# ---------------------
# TF-IDF vectorizer
# ---------------------
tfidf = TfidfVectorizer(max_features=20000, stop_words='english')
X_train_tfidf = tfidf.fit_transform(train_df['clean_text'])
X_test_tfidf = tfidf.transform(test_df['clean_text'])

# ---------------------
# Combine features
# ---------------------
X_train_num = train_df[numeric_cols].fillna(0).values
X_test_num = test_df[numeric_cols].fillna(0).values

X_train_comb = hstack([X_train_tfidf, X_train_num])
X_test_comb = hstack([X_test_tfidf, X_test_num])

# ---------------------
# Train XGBoost
# ---------------------
xgb = XGBClassifier(
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric='logloss'
)
xgb.fit(X_train_comb, train_df['label'])

# ---------------------
# Save everything
# ---------------------
joblib.dump(tfidf, "/content/tfidf_xgb.joblib")
with open("/content/xgb_numeric_cols.json", "w") as f:
    json.dump(numeric_cols, f)
xgb.get_booster().save_model("/content/model_xgb.json")

print("✅ Saved model_xgb.json, tfidf_xgb.joblib, xgb_numeric_cols.json")
```
